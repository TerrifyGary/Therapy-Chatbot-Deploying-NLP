```python
import cloudscraper
import requests
import time
timeout = time.time() +10

articleID = '239238158'

url = "https://www.dcard.tw/service/api/v2/posts/238639330"
```

```python
response = requests.get(url)
print(response.status_code)
```

```python
scraper = cloudscraper.CloudScraper()  # CloudScraper inherits from requests.Session
result = scraper.get(url).text
print(result)
```

```python
isTarget = False
while (isTarget==False):    
    scraper = cloudscraper.CloudScraper()  # CloudScraper inherits from requests.Session
    result = scraper.get(url).text
    if result[0] == '{':
        isTarget = True
    if time.time() > timeout:
        result = "Article Not Found"
        break

print(result)
```

```python
data = result.replace('{','')
data = data.replace('}','')
data = data.replace('"','')
output = data.split(',')

output
```

```python
content = ''
for x in output:
    if x[:7] == 'content':
        content = x.split(':')[1]
        
print(content)
```
