```python
import sqlite3
import pandas as pd
# Create your connection.
cnx = sqlite3.connect('Data/articleList.db')

df = pd.read_sql_query("SELECT * FROM articles", cnx)
```

```python
length = df.shape[0]

aritcleFiltered = []

for x in range (length):
    if (df['withImages'][x] == 0 and df['withVideos'][x]==0 and df['topic'][x]=='mood'):
        aritcleFiltered.append(df['id'][x])
```

```python
from bs4 import BeautifulSoup
import requests

def scapringArticleIndex(id):
    url = "https://www.dcard.tw/f/mood/p/"+id
    r = requests.get(url)
    if r.status_code != 200:
        print('Status Code Error')
        text = 'Status Code Error'
    else:
        soup = BeautifulSoup(r.text,'html.parser')
        filter_1 = soup.find_all('div',class_='sc-4dc2ee70-0 fhYDEv')
        text = filter_1[0].text
        
    return text
```

```python
allArticleIndex = []
for x in range(50):
    print(x)
    allArticleIndex.append(scapringArticleIndex(aritcleFiltered[x]))
```
